## 문제 3 — bank (난이도: 중)

```python
# df불러오기
df2 = pd.read_csv('bank.csv')
df2
```

```python
# Q3-1: balance(통장잔고)가 0 이상인 사람만 필터링해서 평균 나이 구하기
# (age 컬럼은 없지만 duration으로 대체 가능)
df2[df2['balance'] >= 0]['duration'].mean()
df2[df2['balance'] >+ 0]['duration'].mean()
```

```python
# Q3-2: education(교육수준)별 balance 평균 구해서 내림차순 정렬
## 여기서부터 해야한다...
df2.groupby('education')['balance'].mean().sort_values(ascending=False)
# 기준은 에듀케이션,   그리고 밸런스의 평균,    그리고 내림차순 정렬
```

```python
# Q3-3: y(정기예금 가입여부)별 balance 평균을 막대그래프로
df2.groupby('y')['balance'].mean().plot(kind='bar')
```
